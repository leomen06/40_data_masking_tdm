### <a name='mark_01'></a> Introducción

### [Index](#index)


El enmascaramiento de datos (data masking) es una técnica utilizada para proteger la privacidad y seguridad de la información sensible en una base de datos al reemplazar, ocultar o transformar ciertos 
elementos de los datos, mientras se mantiene la integridad y utilidad de los mismos. 

Explicación paso a paso sobre cómo funciona el proceso de data masking:

1. **Identificación de Datos Sensibles:**
   - El primer paso implica identificar los datos en la base de datos que se consideran sensibles y que necesitan ser protegidos. 
   - Estos podrían incluir información personal como nombres, direcciones, números de seguridad social, números de tarjetas de crédito, entre otros.

2. **Definición de Políticas de Enmascaramiento:**
   - Se establecen políticas y reglas para determinar cómo se debe enmascarar cada tipo de dato sensible. 
   - Por ejemplo, se podría decidir ocultar los últimos dígitos de un número de tarjeta de crédito o cambiar aleatoriamente los caracteres de un nombre.
   
   En el contexto del enmascaramiento de datos, los términos "política", "reglas" y "técnicas" a menudo se utilizan de la siguiente manera:

1. **Política de Enmascaramiento de Datos:**
   - **Definición:** La política de enmascaramiento de datos establece los principios y las pautas generales que una organización seguirá en relación con la protección y el enmascaramiento de datos.
   - **Propósito:** Proporciona una visión general de los objetivos y las restricciones asociadas con el manejo de datos confidenciales.

2. **Reglas de Enmascaramiento de Datos:**
   - **Definición:** Las reglas son directrices más específicas y detalladas derivadas de la política. Estas reglas definen cómo se deben aplicar las técnicas de enmascaramiento en situaciones particulares.
   - **Propósito:** Proporciona instrucciones detalladas sobre cómo manejar datos específicos, teniendo en cuenta la naturaleza de esos datos y los requisitos de privacidad.

3. **Técnicas de Enmascaramiento de Datos:**
   - **Definición:** Las técnicas son los métodos específicos utilizados para transformar o enmascarar datos de acuerdo con las reglas establecidas. Pueden incluir métodos como sustitución, truncamiento, encriptación, anonimización, entre otros.
   - **Propósito:** Son las implementaciones prácticas que se aplican para cumplir con las reglas y, por ende, con la política. Cada técnica tiene sus propias características y consideraciones de seguridad.

En resumen, la política establece los principios generales, las reglas proporcionan directrices específicas basadas en la política, y las técnicas son las herramientas o métodos concretos utilizados para aplicar esas reglas y enmascarar los datos de acuerdo con la política establecida. Todas estas capas trabajan juntas para garantizar la protección efectiva de los datos sensibles.

3. **Selección de Técnicas de Enmascaramiento:**
   - Se eligen las técnicas de enmascaramiento adecuadas según las políticas definidas. Algunas técnicas comunes incluyen:
      - **Sustitución:** Reemplazar datos reales con datos ficticios o generados aleatoriamente, no necesariamente consistentes.
      - **Truncamiento:** Reducir la longitud de ciertos campos de datos.
      - **Encriptación:** Convertir datos en un formato cifrado que solo puede ser descifrado por usuarios autorizados.
      - **Anonimización:** Reemplazar datos reales con datos ficticios o generados aleatoriamente, pero consistentes.

4. **Implementación del Proceso de Enmascaramiento:**
   - Se implementa el proceso de enmascaramiento en la base de datos. Esto podría implicar el desarrollo de scripts o la utilización de herramientas especializadas para aplicar las transformaciones de datos según las reglas establecidas.

5. **Validación y Pruebas:**
   - Se realizan pruebas exhaustivas para asegurarse de que el proceso de enmascaramiento no ha afectado la integridad de los datos ni ha introducido problemas en la aplicación que utiliza la base de datos.

6. **Monitoreo Continuo:**
   - Se establece un proceso de monitoreo continuo para garantizar que el enmascaramiento de datos siga siendo efectivo y cumpla con los requisitos de privacidad y seguridad a lo largo del tiempo.

7. **Auditoría y Registro:**
   - Se generan registros y auditorías para rastrear las actividades relacionadas con el enmascaramiento de datos. Esto es crucial para cumplir con regulaciones y normativas de privacidad.

8. **Gestión de Claves y Acceso:**
   - Se implementa una gestión sólida de claves para garantizar que solo las personas autorizadas tengan acceso a los datos enmascarados. Se controla el acceso a las claves de descifrado.

9. **Actualizaciones y Mantenimiento:**
   - Se realiza un seguimiento constante de las políticas de enmascaramiento y se actualizan según sea necesario. También se implementan cambios en las técnicas de enmascaramiento según evolucionan las amenazas y regulaciones.

10. **Cumplimiento Normativo:**
    - Se garantiza que el enmascaramiento de datos cumple con las normativas y regulaciones específicas de la industria y la jurisdicción en la que opera la organización.

La implementación exitosa del enmascaramiento de datos requiere una comprensión detallada de la estructura de la base de datos, las necesidades de privacidad y seguridad, así como un enfoque cuidadoso para garantizar que la utilidad de los datos se mantenga sin comprometer la privacidad.



La TDM (Transformación de Datos Maestros) y el enmascaramiento de datos son conceptos relacionados pero distintos en el ámbito de la gestión de datos. Aquí hay una breve explicación de ambos conceptos y cómo están relacionados:

1. **Transformación de Datos Maestros (TDM):**
   - La TDM se refiere a la práctica de modificar o transformar datos maestros dentro de una organización. Los datos maestros son conjuntos de datos críticos y fundamentales que se utilizan en toda la empresa, como información de clientes, productos o empleados. La TDM puede incluir la limpieza, enriquecimiento y transformación de estos datos para mejorar su calidad y utilidad en los procesos comerciales.

2. **Enmascaramiento de Datos:**
   - El enmascaramiento de datos, como se discutió anteriormente, es el proceso de proteger la privacidad y la seguridad de datos sensibles al reemplazar, ocultar o transformar ciertos elementos de los datos. Se utiliza para garantizar que, incluso en entornos de desarrollo, pruebas o capacitación, los datos sensibles no estén expuestos a usuarios no autorizados.

**Relación:**
   - La relación entre TDM y enmascaramiento de datos radica en que ambos abordan la manipulación de datos para ciertos propósitos. En un entorno de desarrollo o pruebas, por ejemplo, puede ser necesario aplicar tanto técnicas de TDM como de enmascaramiento de datos.

   - Cuando se realiza la TDM en datos maestros, es posible que se necesite enmascarar ciertos atributos o información sensible para cumplir con requisitos de privacidad y seguridad. La combinación de ambas prácticas puede ayudar a garantizar que los datos se utilicen de manera efectiva en diferentes contextos sin comprometer la privacidad.

   - Ambas prácticas son importantes en el contexto de la gestión de datos y la garantía de la calidad de los mismos. La TDM se centra en la calidad y utilidad de los datos maestros, mientras que el enmascaramiento de datos se centra en la privacidad y seguridad de la información sensible.

En resumen, aunque TDM y enmascaramiento de datos tienen objetivos específicos, pueden complementarse mutuamente para garantizar que los datos se gestionen de manera eficaz y segura en toda la organización.


La Transformación de Datos Maestros (TDM) es un proceso que implica la modificación y mejora de los datos maestros en una organización. Aquí hay una explicación paso a paso de cómo se realiza la TDM:

1. **Identificación de Datos Maestros:**
   - Identificar los conjuntos de datos que se consideran datos maestros. Estos son datos fundamentales que se utilizan en toda la organización, como información de clientes, productos, proveedores, empleados, etc.

2. **Análisis de Calidad de Datos:**
   - Realizar un análisis exhaustivo de la calidad de los datos maestros. Esto implica verificar la precisión, integridad, consistencia y actualidad de los datos. Identificar y corregir problemas como duplicados, valores faltantes o incoherencias.

3. **Limpieza de Datos:**
   - Realizar la limpieza de datos para corregir cualquier error o inconsistencia. Esto puede incluir la estandarización de formatos, la corrección de errores tipográficos y la normalización de datos.

4. **Enriquecimiento de Datos:**
   - Enriquecer los datos maestros agregando información adicional de fuentes externas. Esto puede incluir datos demográficos, información de geolocalización u otros detalles relevantes para mejorar la calidad y utilidad de los datos.

5. **Estandarización:**
   - Estandarizar la estructura y el formato de los datos maestros para garantizar la coherencia en su uso. Esto puede implicar la normalización de nombres, direcciones y otros atributos.

6. **Desduplicación:**
   - Identificar y eliminar duplicados en los datos maestros. La presencia de registros duplicados puede afectar la calidad de los informes y análisis.

7. **Gestión de la Calidad Continua:**
   - Establecer procesos y controles para garantizar la calidad continua de los datos maestros. Esto puede incluir la implementación de políticas de entrada de datos, la monitorización regular y la revisión periódica de la calidad de los datos.

8. **Documentación y Metadatos:**
   - Documentar los cambios realizados durante el proceso de TDM. Mantener registros detallados de las transformaciones aplicadas y la lógica utilizada. Los metadatos son esenciales para comprender y auditar los cambios en los datos maestros.

9. **Integración con Sistemas:**
   - Integrar los datos maestros mejorados y transformados en los sistemas de la organización. Asegurarse de que los datos estén disponibles para su uso en diferentes aplicaciones y procesos.

10. **Monitoreo y Mejora Continua:**
    - Establecer un sistema de monitoreo continuo para asegurar la calidad y la eficacia de los datos maestros. Realizar revisiones periódicas y ajustes según sea necesario.

La Transformación de Datos Maestros es un proceso continuo que busca mantener la calidad y relevancia de los datos maestros a lo largo del tiempo. Este enfoque mejora la toma de decisiones, la eficiencia operativa y la capacidad de adaptación de la organización.

### Ejemplo Oracle:

#### Definición de reglas de enmascaramiento.

- Se crea una regla eligiendo alguna de las siguientes opciones de "Masking Rules":

    - Assigment expression
    - Blur Functions
    - De-Identity or Re-Identity Functions
    - Encoding or Decoding Functions
    - **Encryption or Decryption Functions**
    - Generalization via Bucketing
    - Hashing Functions
    - **Pseudonym replaced**
    - **Redaction Function**
    - String Manipulation Functions
    
- Para **Encryption or Decryption Functions** se elige **enc_fp_aes256_alphanum(field,passphrase,exclude)**, para tarjetas de crédito o emails, el cual permite encriptar los datos sin modificar su formato, con esto no corremos el riesgo de corromper datos.
- Para **Pseudonym replaced**, hay que tener un archivo de pseudonimos que se apliquen al campo, name o last_name.

- Para **Redaction Function**, esto funciona colocando `****` que enmascaran los datos.

### Las opciones de connection, para este ejemplo toman en loyalty_src, deployando en loyalty_tgt (Ambiente de testing)

### Las opciones disponibles para el "Target" suelen estar relacionadas con la manipulación y preparación de datos en el entorno de prueba. Aquí están las cuatro especificaciones que has mencionado:

![](img_11.png)
**---------------------------------------------------------------------------------------------------------------------------------------------------------------------**

Truncate table:

Explicación: Esta opción indica que la operación en el destino implica truncar (eliminar todos los registros) de la tabla especificada.
Uso típico: Útil cuando se quiere comenzar con una tabla vacía en el entorno de prueba antes de cargar nuevos datos.

Disable indexes:

Explicación: Indica que se deben desactivar (o deshabilitar) los índices en la tabla de destino antes de realizar operaciones de carga o manipulación.

Uso típico: Puede ser útil deshabilitar índices antes de realizar grandes operaciones de carga para mejorar el rendimiento, ya que la actualización de índices puede ser costosa.

Disable constraints:

Explicación: Esta opción implica desactivar (o deshabilitar) las restricciones (constraints) en la tabla de destino antes de realizar operaciones de carga o manipulación.

Uso típico: Al igual que con los índices, desactivar restricciones puede ser beneficioso al realizar operaciones que podrían violar temporalmente las restricciones, permitiendo así la carga o manipulación de datos.

**Al deshabilitar las restricciones con la opción "Disable constraints", se eliminan temporalmente las restricciones de clave primaria (PK), clave foránea (FK) u otras restricciones definidas en la tabla. Esto puede afectar la integridad de los datos y el diseño del esquema.**

**Aquí hay algunas consideraciones adicionales:**

**Integridad de datos: Al deshabilitar restricciones, la base de datos puede permitir la inserción de datos que de otra manera violarían las restricciones. Por ejemplo, podrían insertarse filas que no cumplan con las restricciones de clave foránea.**

**Diferencias en el esquema: Como mencionas, al deshabilitar restricciones, el esquema de la tabla se altera temporalmente. Esto significa que, durante el tiempo en que las restricciones están deshabilitadas, la tabla puede tener un diseño que difiere del diseño original.**

**Riesgo de datos inconsistentes: Si se realizan operaciones que violan las restricciones mientras están deshabilitadas, podrías terminar con datos inconsistentes o en un estado que no refleja la realidad del entorno de producción.**

**Por lo tanto, al utilizar la opción "Disable constraints", es crucial comprender las implicaciones y, si es necesario, realizar las acciones de limpieza y restauración de las restricciones después de completar las operaciones en la tabla. Esto garantizará que el entorno de prueba vuelva a un estado coherente y consistente con el entorno de producción**

Optimize Relationship for Entities:

Explicación: Esta especificación puede implicar optimizar la relación entre entidades (tablas) para mejorar el rendimiento de las consultas o manipulaciones de datos relacionados.

Uso típico: Puede ser útil al trabajar con datos relacionados en el entorno de prueba para garantizar un rendimiento óptimo durante las pruebas.

Estas opciones permiten configurar cómo se deben manipular las tablas y los datos relacionados en el entorno de prueba, asegurando que estén en el estado adecuado antes y después de la ejecución del plan de Test Data Management.
